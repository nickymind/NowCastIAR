```python
import pandas as pd

# leer insumos
emae = pd.read_csv("../data/emae_yoy.csv")
ipi  = pd.read_csv("../data/ipi_yoy.csv")
isac = pd.read_csv("../data/isac_yoy.csv")

# clave mensual
for d in [emae, ipi, isac]:
    d["date"] = pd.to_datetime(d["date"], errors="coerce")
    d["ym"] = d["date"].dt.strftime("%Y-%m")

# merge SOLO sobre rango com√∫n
common = set(emae["ym"]) & set(ipi["ym"]) & set(isac["ym"])

df = (
    emae[emae["ym"].isin(common)][["ym", "emae_yoy"]]
        .merge(ipi[["ym", "ipi_yoy"]], on="ym", how="left")
        .merge(isac[["ym", "isac_yoy"]], on="ym", how="left")
        .sort_values("ym")
)

# chequeos inmediatos
print(df.tail(12))
print("\nNaN share:")
print(df.isna().mean())

# guardar y volver a leer
df.to_csv("../data/feature_matrix_v1.csv", index=False)
df_check = pd.read_csv("../data/feature_matrix_v1.csv")

print("\nRe-read check:")
print(df_check.tail(12))
```
